## Matplotlib Sample Workflow

```python
# 1. import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt

# 2. Prepare data
x = [1,2,3,4]
y = [12,5,42, 100]

# 3. Setup plot
fig, ax = plt.subplots(figsize=(10,10))

# 4. Plot data
ax.plot(x,y)

# 5. Customize plot
ax.set(title="Simple Plot", xlabel="x-axis", ylabel="y-axis")

# 6. Save and show figure
fig.savefig("sample_plot.png")
```

```python
# Line Plot
fig, ax = plt.subplots()
ax.plot(x, x**2)
```

```python
# Bar Plot
prices = {'Almonds': 10, 'Peanuts': 8, 'Cashews':12}
fig, ax = plt.subplots()
ax.bar(prices.keys(), prices.values())
ax.set(title="Nut Prices", ylabel="Price ($)", xlabel="Nut");
```

```python
fig, ax = plt.subplots()
ax.barh(list(prices.keys()), list(prices.values()))
ax.set(title="Nut Prices", xlabel="Price ($)", ylabel="Nut");
```

```python
# Histogram
x = np.random.randn(1000) # 1000 samples from normal dist.
fig, ax = plt.subplots()
ax.hist(x);
```

```python
# Subplot Option 1
fig, ((ax1,ax2), (ax3,ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10,5))

# fill subplots
ax1.plot(x, x/2)
ax2.scatter(np.random.randint(10, size=10), np.random.randint(10, size=10))
ax3.bar(prices.keys(), prices.values())
ax4.hist(np.random.randn(1000));
```

```python
# Subplot Option 2 (indexing)
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10,5))

ax[0,0].plot(x, x/2)
ax[0,1].scatter(np.random.random(10), np.random.random(10))
ax[1,0].bar(prices.keys(), prices.values())
ax[1,1].hist(np.random.randn(1000));
```

## Pyplot vs Matplotlib Object Oriented
- use pyplot for quick visualization
- use matplotlib for more advanced visualization

```python
disease_data.head()
```

```python
over_50 = disease_data[disease_data["age"] > 50]
```

```python
# PyPlot Method
# c argument specifies color value?
over_50.plot(kind="scatter", x="age", y="chol", c="target");
```

```python
# Object Oriented Method (Matplotlib)
fix, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter", x="age", y="chol", c="target", ax=ax)

# set boundaries for x axis
ax.set_xlim([45,100])
```

```python
# Set up plot
fig, ax = plt.subplots(figsize=(10,6))

# Plot data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])

# Customize plot
ax.set(title="Heart Disease", ylabel="Cholesterol", xlabel="Age")

# Set legend
ax.legend(*scatter.legend_elements(), title="Target")

# Add mean line
ax.axhline(over_50["chol"].mean(), linestyle="--");
```

```python
# Setup plot
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10,10))

# Add data to ax0
scatter = ax0.scatter(x=over_50["age"], 
                      y=over_50["chol"], 
                      c=over_50["target"])

# Customize ax0
ax0.set(title="Heat Disease - Cholersterol", 
        xlabel="Age", 
        ylabel="Cholesterol")

# Add legend
ax0.legend(*scatter.legend_elements(), title="Target")

# Add mean line
ax0.axhline(over_50["chol"].mean(), linestyle="--");

#ax1 Process
scatter1 = ax1.scatter(x=over_50["age"],
                       y=over_50["thalach"], 
                       c=over_50["target"])
ax1.set(title="Heart Disease - Max Heart Rate", 
        xlabel="Age", 
        ylabel="Max Heart Rate")
ax1.legend(*scatter.legend_elements(), title="Target")
ax1.axhline(over_50["thalach"].mean(), linestyle="--");

# Title Entire Figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold")
```

```python
plt.style.use("seaborn-white")
# Setup plot
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10,10))

# Add data to ax0
scatter = ax0.scatter(x=over_50["age"], 
                      y=over_50["chol"], 
                      c=over_50["target"],
                      cmap="winter")

# Customize ax0
ax0.set(title="Heat Disease - Cholersterol", 
        xlabel="Age", 
        ylabel="Cholesterol")
ax0.set_xlim([50,80])

# Add legend
ax0.legend(*scatter.legend_elements(), title="Target")

# Add mean line
ax0.axhline(over_50["chol"].mean(), linestyle="--", c="red");

#ax1 Process
scatter1 = ax1.scatter(x=over_50["age"],
                       y=over_50["thalach"], 
                       c=over_50["target"],
                       cmap="winter")
ax1.set_xlim([50,80])
ax1.set(title="Heart Disease - Max Heart Rate", 
        xlabel="Age", 
        ylabel="Max Heart Rate")
ax1.legend(*scatter.legend_elements(), title="Target")
ax1.axhline(over_50["thalach"].mean(), linestyle="--", c="red");

# Title Entire Figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold")
```
